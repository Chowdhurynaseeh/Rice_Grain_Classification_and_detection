```python
import os
from PIL import Image
from tqdm import tqdm
```

```python
def preprocess_and_save_images(data_dir, output_dir, img_size=(224, 224)):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    labels = os.listdir(data_dir)

    for label in labels:
        input_folder = os.path.join(data_dir, label)
        output_folder = os.path.join(output_dir, label)

        if not os.path.exists(output_folder):
            os.makedirs(output_folder)

        # Get list of images in the current folder
        img_list = os.listdir(input_folder)

        # Use tqdm to show progress for each folder
        for img_name in tqdm(img_list, desc=f"Processing {label}", unit="image"):
            img_path = os.path.join(input_folder, img_name)
            try:
                # Open, resize, and save the image
                img = Image.open(img_path).convert('RGB')
                img = img.resize(img_size)
                img.save(os.path.join(output_folder, img_name))
            except Exception as e:
                print(f"Error processing {img_path}: {e}")
```

```python
data_dir = 'I:/RICE/Rice dataset/Rice Dataset duo'
output_dir = 'I:/RICE/rice_processed_dataset_duo'
```

```python
preprocess_and_save_images(data_dir, output_dir)
```
